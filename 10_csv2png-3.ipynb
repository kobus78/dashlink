```python
#get latest list of hand-labeled non file names from png3/png33/png4
#(single source of truth for handlabels)
nonpath = datpath/labpath/'non'; print(nonpath)
allNonFiles = glob.glob(str(nonpath/'*.png')); #print(allNonFiles)
#allNonFiles = [f.split('/')[3] for f in allNonFiles]
allNonFiles = [f.rsplit('/', 1)[1] for f in allNonFiles]
n_non = len(allNonFiles)
print(f'TOTAL non FILES TO PROCESS: {n_non}\n')
print(allNonFiles)
```

```python
#get latest list of hand-labeled typ file names from png3/png33/png4
#(single source of truth for handlabels)
typpath = datpath/labpath/'typ'; print(typpath)
allTypFiles = glob.glob(str(typpath/'*.png'))
#allTypFiles = [f.split('/')[3] for f in allTypFiles]
allTypFiles = [f.rsplit('/', 1)[1] for f in allTypFiles]
n_typ = len(allTypFiles)
print(f'TOTAL typ FILES TO PROCESS: {n_typ}\n')
print(allTypFiles)
```

```python
#takes about 15 mins on colab
#depending on whether filename is in list of typ files:
#mv all output files (.png) from data dir to png3x/typ|non
#! mv {datpath}/*.png {datpath}/{outpath/typ|non}
allFiles = glob.glob(str(datpath/'*.png'))
n = len(allFiles)
print(f'TOTAL FILES TO MOVE: {n}\n')
for i,file_ in enumerate(allFiles):
    #if i==2: break
    #print(f'file_: {file_}')
    #filename = file_.split('/')[1]; print(f'filename={filename}')
    filename = file_.rsplit('/', 1)[1]; print(f'filename={filename}')
    if filename in allTypFiles:
        ! mv "{file_}" "{datpath}"/{outpath}/typ
    elif filename in allNonFiles:
        ! mv "{file_}" "{datpath}"/{outpath}/non
    else:
        print(f'file {filename} has not been found in either typ or non - it has not been moved!')
    #print(f'{i+1} of {n} complete ........................\n')
    ##print('.')
```
