## Build index

```python
df = pd.read_csv(f'{TAIL}/{MATFILENAME}.csv', 
                     #skiprows=skip,
                     parse_dates={'datetime': ['DATE_YEAR','DATE_MONTH','DATE_DAY','GMT_HOUR','GMT_MINUTE','GMT_SEC']},
                     index_col='datetime', 
                     #usecols=['DATE_YEAR','DATE_MONTH','DATE_DAY','GMT_HOUR','GMT_MINUTE','GMT_SEC',
                     #         'ALT','RALT','BAL1','BAL2',
                     #         'LONP','LATP','DWPT',
                     #         'PTCH','ROLL',
                     #         'CRSS','HDGS','TH','AIL_1','AIL_2','RUDD','TRK'],
                     low_memory=False,
                     #nrows=200000
                     #nrows=3000000
                    )
print(df.shape)
df.head()
```

```python
# df.DATE_YEAR
```

```python
#change index dtype from string to datetime
from datetime import datetime
df.index = [datetime.strptime(s, '%Y %m %d %H %M %S') for s in df.index]
df.index.name = 'datetime'
df = df.drop(['Unnamed: 0'], axis=1)
```
